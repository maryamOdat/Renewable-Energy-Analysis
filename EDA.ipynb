```python
start_year = 1985
end_year = 2021
df_filtered = df[df['Year'].between(start_year, end_year)]

#Group the filtered data by 'Year' and calculate the sum of each resource for each year
summed_df = df_filtered.groupby('Year').sum()

#Rename the columns
summed_df.rename(columns={
    'Electricity_from_wind_GWh': 'Wind',
    'Electricity_from_solar_GWh': 'Solar',
    'Electricity_from_hydro_GWh': 'Hydro',
    'Other_renewables_including_bioenergy_GWh': 'Bioenergy',
    'Renewables_percentage_electricity': 'Renewable',
    'Primary_energy_consumption_TWh': 'Consumption'
}, inplace=True)

#Plotting the total energy consumption and total renewable energy production over the years as a bar chart
plt.figure(figsize=(10, 6))
plt.bar(summed_df.index, summed_df['Consumption'], label='Total Energy Consumption', alpha=0.7)
plt.bar(summed_df.index, summed_df['Renewable'], label='Total Renewable Energy Production', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Energy (TWh)')
plt.title('Total Energy Consumption vs. Total Renewable Energy Production')
plt.legend()
plt.tight_layout()
plt.show()
```



---



<h3>From the Bar chart, we can have below observations:

The bar chart comparison between Total Energy Consumption and Total Renewable Energy Production from 1985 to 2021 shows that renewable energy production has consistently been less than total energy consumption,can be attributed to several factors:
1. Dependency on Fossil Fuels:

    <h5>The global energy landscape has been historically dominated by the use of fossil fuels, such as coal, oil, and natural gas, for electricity generation, transportation, and industrial processes.
Fossil fuels have been the primary source of energy due to their availability, low cost, and established infrastructure.
<br>
2. Limited Renewable Energy Infrastructure:

    <h5>While renewable energy sources, such as solar, wind, hydro, and biomass, have been growing in capacity and adoption, their infrastructure is still not as extensive as that of fossil fuels.
    Building and expanding renewable energy infrastructure, such as solar farms, wind turbines, and hydroelectric plants, takes time, investments, and regulatory approvals.
    <br>
3. Storage and Grid Integration Challenges:

    <h5>Renewable energy sources are often intermittent, depending on factors like weather conditions (e.g., sunlight for solar energy or wind for wind energy).
    Integrating renewable energy into the existing power grid and developing efficient energy storage systems remain challenges.
    <br>
4. Inertia in Energy Systems:

    <h5>Energy systems, including power plants and infrastructure, are long-term investments with a long lifespan.
    Transitioning from traditional energy sources to renewable energy requires significant policy support, public awareness, and industry buy-in.

**Trends in Electricity Generation from Different Renewable Sources by Continent**

```python
#Facetted Line Plot

last_five_years = df['Year'].unique()[-5:]


df_last_five_years = df[df['Year'].isin(last_five_years)]


colors = {'Wind': 'blue', 'Hydro': 'green', 'Solar': 'orange', 'Bioenergy': 'purple'}


f = sns.FacetGrid(df_last_five_years, col='Continent', col_wrap=3, height=4, aspect=1.5)

f.map_dataframe(sns.lineplot, x='Year', y='Electricity_from_wind_GWh', color=colors['Wind'], label='Wind')
f.map_dataframe(sns.lineplot, x='Year', y='Electricity_from_hydro_GWh', color=colors['Hydro'], label='Hydro')
f.map_dataframe(sns.lineplot, x='Year', y='Electricity_from_solar_GWh', color=colors['Solar'], label='Solar')
f.map_dataframe(sns.lineplot, x='Year', y='Other_renewables_including_bioenergy_GWh', color=colors['Bioenergy'], label='Bioenergy')

f.set_axis_labels('Year', 'Electricity Generation (GWh)')


f.fig.suptitle('Trends in Electricity Generation from Different Renewable Sources by Continent')


for ax in f.axes.flat:
    ax.legend()

plt.tight_layout()
plt.show()
```
