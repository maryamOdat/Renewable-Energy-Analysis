- Entity : Refers to the country or region for which the renewable energy data is reported
- Code: A three-letter code assigned to each country or region in the dataset.
- Year: The recorded data corresponds to the specific year of measurement.
- GDP: stands for Gross Domestic Product, and it is a key indicator used in economics to measure the size and growth of an economy. It represents the total value of all goods and services produced within a country's borders during a specific period, typically a year.
- Country Name: Refers to the country
- Country Code: A three-letter code assigned to each country or region in the dataset.
- Electricity from wind (TWh): Total electricity generation in Terawatt-hours (TWh) from wind sources.
- Electricity from hydro (TWh): Total electricity generation in Terawatt-hours (TWh) from hydropower sources.
- Electricity from solar (TWh): Total electricity generation in Terawatt-hours (TWh) from solar sources.

```python
query = '''
Select Year, SUM(Electricity_from_wind_GWh)  as Wind  ,
SUM(Electricity_from_solar_GWh) AS Solar ,
SUM(Electricity_from_hydro_GWh) AS hydro ,
SUM(Other_renewables_including_bioenergy_GWh) AS Bioenergy
FROM df
GROUP BY Year
'''



global_product_year = sqldf(query)
```
